```python
from bs4 import BeautifulSoup

page = requests.get("https://en.wikipedia.org/wiki/Web_scraping")
soup = BeautifulSoup(page.content, 'html.parser')
title = soup.title.text # gets you the text of the <title>(...)</title>
title
```

```python
#Extract page title
page_title = soup.title.text
page_title
```

```python
#Extract body page and head of page
page_body = soup.body
page_head = soup.head
print(page_body, page_head)
```

```python
#Now extract page page of title
page_title = soup.title
page_head = soup.head
print(page_title, page_head)
```

```python
#Now extract the H2 tage
second_h2 = soup.select("h2")[0].text
second_h2
```

## now extract all h2 tag

```python
#now extract all h2 tag
all_h2_tags = []
#set all h2 tag to all h2 tag of the soup
for element in soup.select('h2'):
    all_h2_tags.append(element.text)
print(all_h2_tags)
```

## Now Extract all P tag

```python
all_p_tags = []
#set all P tag to all p tag of the soup
for element in soup.select('p'):
    all_p_tags.append(element.text)
print(all_p_tags)
```
